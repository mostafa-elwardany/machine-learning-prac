## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import pandas as pd

url = "https://raw.githubusercontent.com/mostafa-elwardany/machine-learning-prac/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Number 1: how the pages split by trend direction
trend_split = df["trend_direction"].value_counts(normalize=True)
print("Trend direction split:\n", trend_split)

# Number 2: does position correlate with impressions?
position_compare = df.groupby(df["avg_position"] <= 10)["impressions_90d"].mean()
print("\nMean impressions_90d, position <=10 vs >10:\n", position_compare)

# Number 3: how many pages look "stale but visible" (worth investigating)
stale_visible = df[(df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)]
print(f"\nStale but visible pages: {len(stale_visible)} out of {len(df)} ({len(stale_visible)/len(df):.1%})")
```
